```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Making a list of missing value type
missing_value =['n/a', 'na', '--']
# Read in the data
df = pd.read_csv('C:/abcd/movies.csv', na_values=missing_value)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Replace missing values using median values

median=df['budget'].median()
df['budget'].fillna(median, inplace=True)

median=df['gross'].median()
df['gross'].fillna(median, inplace=True)

median=df['votes'].median()
df['votes'].fillna(median, inplace=True)

median=df['runtime'].median()
df['runtime'].fillna(median, inplace=True)
```

```python
df.head()
```

```python
# Change the dtype of some columns

df['budget']=df['budget'].astype('int64')
df['votes']=df['votes'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
# Budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color': 'blue'})
plt.show()
```

```python
correlation_matrix=df.corr(method='pearson')
```

```python
# Correlation heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matric')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
